```python
import os

reference_dir = r"C:\Users\TIMEA\Desktop\Summarize DatasSet\Reference" 
testdata_dir = r"C:\Users\TIMEA\Desktop\Summarize DatasSet\TestData" 

# save candidate_summary and reference_summary
candidate_summaries = []
reference_summaries = []

tasks = sorted(os.listdir(testdata_dir))  

for task in tasks:
    task_test_dir = os.path.join(testdata_dir, task)  
    task_ref_dir = os.path.join(reference_dir, task) 

    if os.path.isdir(task_test_dir) and os.path.isdir(task_ref_dir):
        test_files = sorted(os.listdir(task_test_dir))  
        ref_files = sorted(os.listdir(task_ref_dir))  

        if len(test_files) != len(ref_files):
            print(f"warn：{task} is not match with reference！")
            continue

        # 
        for i, (test_file, ref_file) in enumerate(zip(test_files, ref_files), start=1):
            test_path = os.path.join(task_test_dir, test_file) 
            ref_path = os.path.join(task_ref_dir, ref_file)  

            # read TestData
            with open(test_path, "r", encoding="utf-8") as f:
                raw_text = f.read().strip()

        
            print(f" generate {test_file} summarization...")
            candidate_summary = generate_summary(raw_text)

         
            with open(ref_path, "r", encoding="utf-8") as ref_file:
                reference_summary = ref_file.read().strip()

     
            candidate_summaries.append(candidate_summary)
            reference_summaries.append(reference_summary)
    else:
        print(f"warn：TestData or Reference of {task}  is not exist")

# ensure candidate and reference have same file type
assert len(candidate_summaries) == len(reference_summaries), "reference and candidate file don't match！"
```

```python
P, R, F1 = score(candidate_summaries, reference_summaries, lang="en", model_type="bert-base-uncased")


for i, (cand, ref) in enumerate(zip(candidate_summaries, reference_summaries)):
    print(f" Test File {i+1}:")
    print(f"  Precision: {P[i].item():.4f}, Recall: {R[i].item():.4f}, F1-score: {F1[i].item():.4f}\n")


print(f" Average F1-score: {F1.mean().item():.4f}")
```

```python
rouge_scorer_fn = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

rouge1_scores = []
rouge2_scores = []
rougeL_scores = []

for cand, ref in zip(candidate_summaries, reference_summaries):
    rouge_scores = rouge_scorer_fn.score(ref, cand)
    
    rouge1_scores.append(rouge_scores['rouge1'].fmeasure)
    rouge2_scores.append(rouge_scores['rouge2'].fmeasure)
    rougeL_scores.append(rouge_scores['rougeL'].fmeasure)

# **Display ROUGE results**
for i in range(len(candidate_summaries)):
    print(f" Test File {i+1}:")
    print(f"  ROUGE-1 F1: {rouge1_scores[i]:.4f}, ROUGE-2 F1: {rouge2_scores[i]:.4f}, ROUGE-L F1: {rougeL_scores[i]:.4f}\n")

# **Compute and print average ROUGE scores**
print(f" Average ROUGE-1 F1-score: {sum(rouge1_scores) / len(rouge1_scores):.4f}")
print(f" Average ROUGE-2 F1-score: {sum(rouge2_scores) / len(rouge2_scores):.4f}")
print(f" Average ROUGE-L F1-score: {sum(rougeL_scores) / len(rougeL_scores):.4f}")
```
