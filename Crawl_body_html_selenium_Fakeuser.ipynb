```python
#2
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

# 设置 Chrome 的无头模式
chrome_options = Options()
chrome_options.add_argument("--headless")  # 无界面模式
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")

# 【绕过 Selenium 检测】
chrome_options.add_argument("--disable-blink-features=AutomationControlled")

# 【添加 User-Agent 伪装成普通浏览器】
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36")

# 【尝试关闭 HTTP/2，避免 ERR_HTTP2_PROTOCOL_ERROR】
chrome_options.add_argument("--disable-features=NetworkService,NetworkServiceInProcess")

chrome_options.binary_location = "/usr/bin/chromium-browser"

# 启动 Selenium WebDriver
driver = webdriver.Chrome(options=chrome_options)

# 访问 Tesla 隐私政策页面
#url = "https://www.tesla.com/legal/privacy"
# 访问google
#url = "https://policies.google.com/privacy?hl=en-US"
#ibm
#url = "https://www.ibm.com/us-en/privacy"
#tiktok
#url = "https://www.tiktok.com/legal/page/row/privacy-policy/en"
#uber
#url = "https://www.uber.com/global/en/privacy/overview/"
#Reddit
#url = "https://www.reddit.com/policies/privacy-policy?__hstc=53109431.bd1028bfa9b412cc559ab7e732b52496.1727735173021.1732200491510.1732209607383.87&__hssc=53109431.4951.1732209607383&__hsfp=810579359y"
#dropbox
#url = "https://www.dropbox.com/en_GB/terms"
#apple
#url = "https://www.apple.com/legal/privacy/en-ww/"
#signal
#url = "https://signal.org/legal/"
#superhuman
url = "https://superhuman.com/privacy"


driver.get(url)

# 等待 JavaScript 生成内容
time.sleep(5)  # Tesla 可能需要更长时间加载 JavaScript

# 获取 <body> 标签及其内部的 HTML 代码
body_html = driver.find_element("tag name", "body").get_attribute("outerHTML")
```

```python
###3
#可视化html，引入 html2text 并进行转换
!pip install selenium html2text
!apt-get update
!apt-get install -y chromium-chromedriver
```

```python
#4
import html2text
# 将 HTML 转换为 Markdown 纯文本
readable_text = html2text.html2text(body_html)
```

```python
###5
### 输出可读文本
print(readable_text)
# 输出完整的 <body> HTML 代码
print(body_html)
```
